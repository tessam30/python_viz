```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
housing_raw = pd.read_csv("../Data/housing_data.csv")
```

```python
ca_housing = (
    housing_raw
    .loc[housing_raw["region_name"].str.contains("CA")]
    .assign(region_name = housing_raw["region_name"].str[0:-11])
    .query("region_name in ['San Francisco', 'Los Angeles']")
)
```

```python
ca_housing_markets = (ca_housing
                     .set_index("period_begin")
                     .loc[:, ["region_name", "median_active_list_price"]]
                     .sort_index()
                     .pivot_table(index="period_begin",
                                  columns="region_name",
                                  values="median_active_list_price")
                     .resample("Q")
                     .mean()
                    )

ca_housing = ca_housing_markets.loc[: , ["San Francisco", "Los Angeles"]]
```

```python
fig, ax = plt.subplots()

ax.plot(
    ca_housing.index, 
    ca_housing["San Francisco"],
    label="San Francisco"
)
ax.plot(
    ca_housing.index, 
    ca_housing["Los Angeles"],
    label="Los Angeles"
)

fig.suptitle("San Francisco Prices Decline in 2020")
ax.set_title("While LA Prices Keep Rising")
ax.set_xlabel("Year")
ax.set_ylabel("Median Home Price (Millions)")

ax.legend(["SF", "LA"])
```

```python
fig, ax = plt.subplots()

ax.plot(
    ca_housing.index, 
    ca_housing["San Francisco"],
    color="green",
    linewidth=2
)
ax.plot(
    ca_housing.index, 
    ca_housing["Los Angeles"],
    color="purple",
    linewidth=2,
    linestyle="--"    
)

fig.suptitle("San Francisco Prices Decline in 2020")
ax.set_title("While LA Prices Keep Rising")
ax.set_xlabel("Year")
ax.set_ylabel("Median Home Price (Millions)")

ax.legend(ca_housing.columns)
```

```python
fig, ax = plt.subplots()

ax.plot(
    ca_housing.index, 
    ca_housing["San Francisco"],
    color="green",
    linewidth=2
)
ax.plot(
    ca_housing.index, 
    ca_housing["Los Angeles"],
    color="purple",
    linewidth=2
)

fig.suptitle("San Francisco Prices Decline in 2020")
ax.set_title("While LA Prices Keep Rising")
ax.set_xlabel("Year")
ax.set_ylabel("Median Home Price (Millions)")

ax.legend(ca_housing.columns)

ax.set_xlim(17500, 19000)
ax.set_ylim(0, 1600000)
```

```python
fig, ax = plt.subplots(figsize=(7, 6))

ax.plot(
    ca_housing.index, 
    ca_housing["San Francisco"],
    color="green",
    linewidth=2
)
ax.plot(
    ca_housing.index, 
    ca_housing["Los Angeles"],
    color="purple",
    linewidth=2
)

fig.suptitle(
    "San Francisco Prices Decline in 2020", 
    y=.95
)
ax.set_title("While LA Prices Keep Rising")
ax.set_xlabel("Year")
ax.set_ylabel("Median Home Price (Millions)")

ax.legend(ca_housing.columns)
ax.set_ylim(0, 1600000)

ax.axvline(18341, c="black", ls="--")
```

```python
fig, ax = plt.subplots(figsize=(7, 6))

fig.suptitle(
    "San Francisco Prices Decline in 2020", 
    y=.95
)

ax.set_title("While LA Prices Keep Rising")
ax.set_xlabel("Year")
ax.set_ylabel("Median Home Price (Millions)")

ax.plot(
    ca_housing.index, 
    ca_housing["San Francisco"], 
    color="green",
    linewidth=2
)

ax.plot(
    ca_housing.index, 
    ca_housing["Los Angeles"], 
    color="purple", 
    linewidth=2
)

ax.legend(ca_housing.columns[::-1])
ax.set_ylim(0, 1600000)
ax.axvline(18341, c="black", ls="--")
ax.annotate("Start of Covid -->",
            xy=(18300, 1000000),
            xytext=(17850, 1000000)
)

ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(7, 6))

ax.plot(
    ca_housing.index, 
    ca_housing["San Francisco"]/1000000,
    color="green",
    linewidth=2
)
ax.plot(
    ca_housing.index, 
    ca_housing["Los Angeles"]/1000000,
    color="purple",
    linewidth=2
)

fig.suptitle(
    "San Francisco Prices Decline in 2020", 
    y=.95
)
ax.set_title("While LA Prices Keep Rising")
ax.set_xlabel("Year")
ax.set_ylabel("Median Home Price (Millions)")

ax.legend(ca_housing.columns)
ax.set_ylim(0, 1.6)

ax.axvline(18341, c="black", ls="--")
ax.annotate("Start of Covid -->",
            xy=(18300, 1),
            xytext=(17850, 1)
)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```

```python
fig, ax = plt.subplots(figsize=(7, 6))

ax.plot(
    ca_housing.index, 
    ca_housing["San Francisco"]/1000000,
    color="green",
    linewidth=2
)
ax.plot(
    ca_housing.index, 
    ca_housing["Los Angeles"]/1000000,
    color="purple",
    linewidth=2
)

fig.text(.15, .9, 
         "San Francisco Prices Hit Hard By Covid...",
         fontsize=14
        )
ax.text(18400, .6, 
         "...and LA prices keep rising",
         fontsize=12
)
ax.set_xlabel("Year")
ax.set_ylabel("Median Home Price (Millions)")

ax.legend(ca_housing.columns)
ax.set_ylim(0, 1.6)

ax.axvline(18341, c="black", ls="--")
ax.annotate("Start of Covid",
            xy=(18300, 1),
            xytext=(17850, 1),
            arrowprops=dict(
                facecolor="black", 
                width=.5, 
                headwidth=6,
                connectionstyle="angle3, angleA=290, angleB=0"
            ),
            verticalalignment="center"
)
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
```
