```python
people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

def split_title_and_name(person):
    title = person.split()[0]
    last = person.split()[-1]
    return '{} {}'.format(title, last)
    
list(map(split_title_and_name, people))    

    
```

```python
people = ['Dr. Christopher Brooks', 'Dr. Kevyn Collins-Thompson', 'Dr. VG Vinod Vydiswaran', 'Dr. Daniel Romero']

def split_title_and_name(person):
    return person.split()[0] + ' ' + person.split()[-1]


```

```python
#option 1
for person in people:
    print(split_title_and_name(person) == (lambda x: x.split()[0] + ' ' + x.split()[-1])(person))
```

```python
names = map(lambda x: x.split()[0] + ' ' + x.split()[-1], people)
print(list(names))
```

```python
# Can collect results in x and append the lambda function results to the empty object

x = []
for person in people:
    x.append((lambda x: x.split()[0] + ' ' + x.split()[-1])(person))
    
x
```

```python
# List comprehension example: translate the following into a list comprehension example
my_list = []
for number in range(0,1000):
    if number % 2 == 0:
        my_list.append(number)
        
my_list        
```

```python
my_list_comp = [number for number in range(1000) if number % 2 == 0]
my_list_comp
```

```python
def times_tables():
    lst = []
    for i in range(10):
        for j in range (10):
            lst.append(i*j)
    return lst

times_tables()
```

```python
times_tables() == [i * j for i in range(10) for j in range(10)]
times_tables()
```
