```python
import matplotlib.pyplot as plt 

# Create a mesh grid with numpy
points = np.arange(-5, 5, 0.1)
xs, ys = np.meshgrid(points, points)

z = np.sqrt(xs ** 2 + ys ** 2)

fig, ax = plt.subplots(figsize = (10, 10))
plt.scatter(xs, ys, s = 1)
```
