```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
import os
import numpy as np
```

```python
diamonds = pd.read_csv("../Data/Diamonds Prices2022.csv")
diamonds.head()
```

```python
sns.barplot(
    x = "cut",
    y = "carat", 
    data = diamonds, 
    hue = "cut"
)
```

```python
sns.barplot(
    y = "cut",
    x = "carat", 
    data = diamonds, 
    hue = "cut",
  #facecolor = "white",
    #edgecolor = "black"
)
```

```python
# Grouped Bar charts
sns.barplot(
    x = "cut",
    y = "carat", 
    hue = "color",
    data = diamonds.sort_values(by = "color"),
    palette = "viridis"
)
```

```python
sns.histplot(
    x = "price",
    data = diamonds,
    kde = True
)
```

```python
# Build seaborn plot around matlplot lib
fig, ax = plt.subplots(figsize = (8, 8))

sns.barplot(
    y = "cut",
    x = "price",
    hue = "clarity", 
    data = diamonds,
    palette = "husl",
    edgecolor = "white"
)

ax.legend(ncols = 8)
```

```python
diamonds_pivot = diamonds.pivot_table(
    index = "cut",
    columns = "color",
    values = "carat",
    aggfunc = "mean"
)
```

```python
sns.heatmap(diamonds_pivot, linewidth = 0.01)
```
